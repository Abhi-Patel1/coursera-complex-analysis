# Drawing Filled-in Julia Set for  $f(z) = z^2+c$

```python
from __future__ import print_function, division
import numpy as np
import pandas as pd
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def f(z, c):
    return z * z + c

def val(v, size, B):
    return (v / (size-1) - 0.5) * B * 2

def calculate_julia_set(c, size, max_iter):
    R = (1 + math.sqrt(1 + 4 * abs(c))) / 2
    B = R * 1.1
    A = np.zeros((size, size), dtype=int)
    for s in xrange(size):
        for t in xrange(size):
            z = val(s, size, B) + val(t, size, B) * 1j
            i = 0
            while True:
                if abs(z) >= R:
                    A[t,s] = i
                    break
                elif i == max_iter - 1:
                    A[t,s] = max_iter
                    break
                else:
                    z = f(z, c)
                    i = i + 1
    return A, B

def draw_julia_set(A, B, max_iter, title_text):
    my_cmap = mpl.cm.get_cmap('OrRd')
    my_cmap.set_under('w')
    my_cmap.set_over('k')

    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    cax = ax.matshow(np.flipud(A), interpolation='none', cmap=my_cmap, vmin=1, vmax=max_iter-1)
    ax.set_xticklabels(['%.2f'%val(s, A.shape[1], B) for s in ax.get_xticks()])
    ax.set_yticklabels(['%.2f'%-val(t, A.shape[1], B) for t in ax.get_yticks()])
    ax.set_title(title_text, fontsize=16)
    fig.colorbar(cax)
    
    plt.show()
```

```python
size = 601
max_iter = 10
c = 0
A, B = calculate_julia_set(c, size, max_iter)
draw_julia_set(A, B, max_iter, 'c = 0')
```
