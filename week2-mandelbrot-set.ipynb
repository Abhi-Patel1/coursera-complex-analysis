# Drawing Mandelbrot Set

```python
from __future__ import print_function, division
import numpy as np
import pandas as pd
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def f(z, c):
    return z * z + c

def val(v, vmin, vmax, size):
    return v / (size-1) * (vmax- vmin) + vmin

def calculate_mandelbrot_set(bb, size, max_iter):
    R = 2
    A = np.zeros((size, size), dtype=int)
    for s in xrange(size):
        for t in xrange(size):
            c = val(s, bb['xmin'], bb['xmax'], size) + val(t, bb['ymin'], bb['ymax'], size) * 1j
            z = 0
            i = 0
            while True:
                if abs(z) >= R:
                    A[t,s] = i
                    break
                elif i == max_iter - 1:
                    A[t,s] = max_iter
                    break
                else:
                    z = f(z, c)
                    i = i + 1
    return A

def draw_mandelbrot_set(bb, A, max_iter):
    my_cmap = mpl.cm.get_cmap('OrRd')
    my_cmap.set_under('w')
    my_cmap.set_over('k')

    fig = plt.figure(figsize=(15,12))
    ax = fig.add_subplot(111)
    cax = ax.matshow(np.flipud(A), interpolation='none', cmap=my_cmap, vmin=1, vmax=max_iter-1)
    ax.set_xticklabels(['%.3f'%val(s, bb['xmin'], bb['xmax'], A.shape[1]) for s in ax.get_xticks()])
    ax.set_yticklabels(['%.3f'%val(t, bb['ymin'], bb['ymax'], A.shape[0]) for t in ax.get_yticks()[::-1]])
    fig.colorbar(cax)
    
    plt.show()
```

```python
size = 801
bb = {'xmin': -2.1, 'xmax': 0.7, 'ymin': -1.4, 'ymax': 1.4}
max_iter = 100
A = calculate_mandelbrot_set(bb, size, max_iter)
draw_mandelbrot_set(bb, A, max_iter)
```
