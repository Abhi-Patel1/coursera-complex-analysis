```python
from __future__ import print_function, division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid.axislines import SubplotZero
```

```python
%matplotlib inline
```

### PROBLEM 1

```python
%%latex
(a) $\{z \in C \mid |z-3-2i| \le 1 \}$
```

```python
fig = plt.figure(1, figsize=[6,6])
ax = SubplotZero(fig, 111)
fig.add_subplot(ax, aspect='equal')

for direction in ["xzero", "yzero"]:
    ax.axis[direction].set_axisline_style("->", size=1.5)
    ax.axis[direction].set_visible(True)
    ax.axis[direction].major_ticklabels.set_fontsize(13)    

for direction in ["left", "right", "bottom", "top"]:
    ax.axis[direction].set_visible(False)

ax.set_xlim(0,4)
ax.set_ylim(0,4)
ax.set_xticks([0,1,2,3,4])
ax.set_yticks([0,1,2,3,4])
    
unit_circle = plt.Circle((3,2), 1, color='r', fill=False, hatch='/', ls='-', lw=2)
ax.add_artist(unit_circle)
ax.annotate('the set', xy=(2.5, 1.5), xytext=(2.1, 0.9), ha='right', va='top', fontsize=14, color='r',
            arrowprops=dict(facecolor='r', edgecolor='r', arrowstyle='simple'))
ax.annotate('boundary\n included\n in the set', xy=(2.15, 2.6), xytext=(2.0, 3.0),
            ha='right', va='bottom', fontsize=14, color='r',
            arrowprops=dict(facecolor='r', edgecolor='r', arrowstyle='simple'))

ax.text(3.6, 0.15, 'x=Re(z)', fontsize=14)
ax.text(0.15, 4.05, 'y=Im(z)', fontsize=14)
    
ax.scatter(3, 2, color='b', s=50)
ax.plot([3,3], [2,0], color='b', ls='--', lw=2)
ax.plot([3,0], [2,2], color='b', ls='--', lw=2)
ax.plot([3,3.8], [2,2.6], color='b', ls='--', lw=2)
    
ax.annotate('radius is 1', xy=(3.4, 2.4), xytext=(3.4, 3.2), ha='center', va='bottom', fontsize=14, color='b',
            arrowprops=dict(facecolor='b', edgecolor='b', arrowstyle='simple'))

plt.show()
```

```python
%%latex
(b) $\{z \in C \mid Im \: z = 2 \}$
```

```python
fig = plt.figure(1, figsize=[6,6])
ax = SubplotZero(fig, 111)
fig.add_subplot(ax, aspect='equal')

for direction in ["xzero", "yzero"]:
    ax.axis[direction].set_axisline_style("->", size=1.5)
    ax.axis[direction].set_visible(True)
    ax.axis[direction].major_ticklabels.set_fontsize(13)    

for direction in ["left", "right", "bottom", "top"]:
    ax.axis[direction].set_visible(False)

ax.set_xlim(-1,3)
ax.set_ylim(-1,3)
ax.set_xticks([-1,0,1,2,3])
ax.set_yticks([-1,0,1,2,3])
    
ax.plot([-1,3], [2,2], color='r', ls='-', lw=2)
ax.annotate('this line is the set', xy=(1, 1.95), xytext=(1, 1.3), ha='center', va='top', fontsize=14, color='r',
            arrowprops=dict(facecolor='r', edgecolor='r', arrowstyle='simple'))

ax.text(2.6, 0.15, 'x=Re(z)', fontsize=14)
ax.text(0.15, 3.05, 'y=Im(z)', fontsize=14)

plt.show()
```

```python
%%latex
(c) $\{z \in C \setminus \{0\} \mid 0 < arg \, z < \pi/6 \}$
```

```python
fig = plt.figure(1, figsize=[6,6])
ax = SubplotZero(fig, 111)
fig.add_subplot(ax, aspect='equal')

for direction in ["xzero", "yzero"]:
    ax.axis[direction].set_axisline_style("->", size=1.5)
    ax.axis[direction].set_visible(True)
    ax.axis[direction].major_ticklabels.set_fontsize(13)    

for direction in ["left", "right", "bottom", "top"]:
    ax.axis[direction].set_visible(False)

ax.set_xlim(-1,3)
ax.set_ylim(-1,3)
ax.set_xticks([-1,0,1,2,3])
ax.set_yticks([-1,0,1,2,3])
    
ax.plot([0,3], [0,0], color='r', ls='--', lw=2)
ax.plot([0,1.5*np.sqrt(3)], [0,1.5], color='r', ls='--', lw=2)
ax.annotate('the set', xy=(2.5, 1), xytext=(2.8, 1.8), ha='center', va='bottom', fontsize=14, color='r',
            arrowprops=dict(facecolor='r', edgecolor='r', arrowstyle='simple'))

ax.text(2.6, 0.15, 'x=Re(z)', fontsize=14)
ax.text(0.15, 3.05, 'y=Im(z)', fontsize=14)

ax.scatter(0,0,edgecolor='r',facecolor='none',marker='o',s=100,lw=2)
ax.annotate('origin\n not included\n in the set', xy=(-0.05, 0.1), xytext=(-0.3, 0.6), ha='center', va='bottom', fontsize=14, color='r',
            arrowprops=dict(facecolor='r', edgecolor='r', arrowstyle='simple'))

angle_wedge = mpl.patches.Wedge([0,0], 3, 0, 30, facecolor='none', edgecolor='r', hatch='/', lw=0)
ax.add_artist(angle_wedge)
ax.annotate('line not included\n in the set', xy=(1.5, 0.95), xytext=(1.25, 1.4), ha='center', va='bottom', fontsize=14, color='r',
            arrowprops=dict(facecolor='r', edgecolor='r', arrowstyle='simple'))
ax.annotate('line not included\n in the set', xy=(1.7, -0.05), xytext=(1.7, -0.55), ha='center', va='top', fontsize=14, color='r',
            arrowprops=dict(facecolor='r', edgecolor='r', arrowstyle='simple'))

angle_arc = mpl.patches.Arc([0,0], 0.5, 0.5, 0, 0, 30, color='b', lw=2)
ax.add_artist(angle_arc)
ax.annotate('$\pi/6$', xy=(0.25, 0.06), xytext=(1.0, 0.24), ha='left', va='bottom', fontsize=16, color='b',
            arrowprops=dict(facecolor='b', edgecolor='b', arrowstyle='simple'))

plt.show()
```

```python
fig = plt.figure(1, figsize=[6,6])
ax = SubplotZero(fig, 111)
fig.add_subplot(ax, aspect='equal')

for direction in ["xzero", "yzero"]:
    ax.axis[direction].set_axisline_style("->", size=1.5)
    ax.axis[direction].set_visible(True)
    ax.axis[direction].major_ticklabels.set_fontsize(13)    

for direction in ["left", "right", "bottom", "top"]:
    ax.axis[direction].set_visible(False)

ax.set_xlim(-1,3)
ax.set_ylim(-1,3)
ax.set_xticks([-1,0,0.5,1,2,3])
ax.set_yticks([-1,0,1,2,3])
    
ax.plot([0.5,0.5], [-1,3], color='r', ls='--', lw=2)
ax.annotate('line not\n included\n in the set', xy=(0.45, 1.5), xytext=(-0.1, 1.5), ha='right', va='center', fontsize=14, color='r',
            arrowprops=dict(facecolor='r', edgecolor='r', arrowstyle='simple'))

ax.text(2.6, 0.15, 'x=Re(z)', fontsize=14)
ax.text(0.15, 3.05, 'y=Im(z)', fontsize=14)

rect = mpl.patches.Rectangle([0.5,-1], 2.5, 4, facecolor='none', edgecolor='r', hatch='/', lw=0)
ax.add_artist(rect)
ax.text(1.7, 1, 'the set', ha='center', va='bottom', fontsize=14, color='r')

plt.show()
```
